# Hist Plot

```python
for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

# Count Plot

```python
# Categorical Features
categorical_features = ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'Inflight wifi service', 
                         'Departure/Arrival time convenient', 'Ease of Online booking', 'Gate location', 
                         'Food and drink', 'Online boarding', 'Seat comfort', 'Inflight entertainment', 
                         'On-board service', 'Leg room service', 'Baggage handling', 'Checkin service', 
                         'Inflight service', 'Cleanliness']
for feature in categorical_features:
    sns.countplot(x=feature, data=df)
    plt.xticks(rotation=90)
    plt.show()
```

```python

# 17. Departure Delay vs. Satisfaction
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Departure Delay in Minutes', y='satisfaction', data=df)
plt.title('Departure Delay vs. Satisfaction')
plt.xlabel('Departure Delay (Minutes)')
plt.ylabel('Satisfaction')
plt.show()

# 18. Arrival Delay vs. Satisfaction
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Arrival Delay in Minutes', y='satisfaction', data=df)
plt.title('Arrival Delay vs. Satisfaction')
plt.xlabel('Arrival Delay (Minutes)')
plt.ylabel('Satisfaction')
plt.show()
```

# Heatmap

```python
# 19. Correlation Heatmap of All Numerical Features
plt.figure(figsize=(12, 10))
correlation_matrix = df[['Age', 'Flight Distance', 'Departure Delay in Minutes', 'Arrival Delay in Minutes']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

# catplot

```python
# 20. Satisfaction by Class and Type of Travel
plt.figure(figsize=(12, 8))
sns.catplot(x='Class', hue='Type of Travel', col='satisfaction', data=df, kind='count', height=6, aspect=1.5)
plt.title('Satisfaction by Class and Type of Travel')
plt.show()
```
